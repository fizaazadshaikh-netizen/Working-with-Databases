```python
import sqlite3

conn = sqlite3.connect('foodie.db')
cursor = conn.cursor()

cursor.execute("create table if not exists restaurants (id integer primary key, name text, cuisine text, rating real)")
conn.commit()
conn.close()
```

```python
import sqlite3

conn = sqlite3.connect('foodie.db')
cursor = conn.cursor()

cursor.executemany("insert into restaurants (id, name, cuisine, rating) values (?, ?, ?, ?)", [
    (1, 'swagat', 'indian', 4.2),
    (2, 'italiano', 'italian', 4.5),
    (3, 'dragon wok', 'chinese', 3.9)
])
conn.commit()

cursor.execute("select name from restaurants where rating > 4.0")
for row in cursor.fetchall():
    print(row[0])

conn.close()
```

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect('foodie.db')
df = pd.read_sql_query("select * from restaurants", conn)
print(df.head(2))
conn.close()
```

```python
df['delivery_charge'] = 50
df['final_rating'] = df.apply(lambda x: x['rating'] + 0.1 if x['cuisine'].lower() == 'italian' else x['rating'], axis=1)
print(df)
```

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect('foodie.db')
df_top = pd.read_sql_query("select * from restaurants where rating > 4.5", conn)
df_top.to_csv('top_rated_restaurants.csv', index=False)
conn.close()
```
